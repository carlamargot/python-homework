Savings Health Analysis 

```python
#monthly average income
monthly_income = 12000

# creating a new df 
savings_df = pd.DataFrame()
# adding values to my new df 
crypto_value = (btc * btc_price) + (eth * eth_price)
shares_value = (my_spy * spy_close) + (my_agg * agg_close)

#formatting 
data = {
     'amount' : [crypto_value, shares_value]
}

port = ['crypto', 'shares']
# setting the index to crypto and shares 
savings_df = pd.DataFrame(data, index= port)
```

```python
savings_df.round().head()
```

```python
# plotting a pie chart
savings_df.plot.pie(y='amount', title = 'Personal Savings Portfolio')
```

```python
# calculating emergency fund 
ideal_Efund = monthly_income * 3
ideal_Efund
```

```python
# seeing if funds are ideal 

def savings(x):
    if x >= ideal_Efund:
        print("Congratulations, you have an ideal emergncy fund!")
    elif x == ideal_Efund:
        print("Congratulations, you have reached your savings goal!")
    else:
        print(f"You are {(ideal_Efund - x)} away from your savings goal!")
```

```python
# testing out the function
savings(37000)
```

```python
load_dotenv()
```

```python
# setting up keys 
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")


# setting up object 
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
# testing
type(alpaca_secret_key)
```

```python
# selecting start and end dates 
start_date = pd.Timestamp('2015-1-18', tz='America/New_york').isoformat()
end_date = pd.Timestamp('2021-1-18', tz='America/New_york').isoformat()
```

```python
# retrieving 5 years of histroical data for AGG and SPY
tickers = ["SPY","AGG"]
timeframe = "1D"

historical_data = alpaca.get_barset(
    tickers, 
    timeframe,
    start = start_date, 
    end = end_date
).df
```

```python
historical_data
```

```python
# monte carlo stimulation of 30 years 
MC_30 =  MCSimulation(
    portfolio_data =historical_data,
    weights = [.60,.40],
    num_simulation = 500,
    num_trading_days = 252 * 30
)
```

```python
# printing stimulation data 
MC_30.portfolio_data.head(10)
```

```python
# running the MC simulation of 30 years cumulative returns 
MC_30.calc_cumulative_return()
```

```python
# plotting MC simulation outcomes 
MC_line_plot = MC_30.plot_simulation()
```

```python
# plotting probablity variables and convidence intervals 
distribution_MC = MC_30.plot_distribution()
```

```python
# summary stastics 
MC_summary = MC_30.summarize_cumulative_return()
MC_summary
```

```python
# Calculate the expected portfolio return at the 95% lower and upper confidence intervals based on a $20,000 initial investment.
initial_investment = 20000

upper_CI = round(MC_summary[9]*20000,2)
lower_CI = round(MC_summary[8]*20000,2)
```

```python
print(upper_CI)
print(lower_CI)
```

```python
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${lower_CI} and ${upper_CI}")
```
